```python
import pandas as pd

# Load the dataset
data = pd.read_csv(r"C:\Users\nadha.NICHU\OneDrive\Desktop\epi_r.csv")

# Step 1: Initial Data Overview
print("Initial Data:\n", data.head())

# Step 2: Check for missing values
print("\nMissing Values:\n", data.isnull().sum())

# Step 3: Handle missing values
data_cleaned = data.dropna(subset=['rating', 'calories', 'protein', 'fat']) 

# Step 4: Check for duplicates
print("\nNumber of Duplicate Entries:", data_cleaned.duplicated().sum())

# Step 5: Remove duplicates if any
data_cleaned = data_cleaned.drop_duplicates()

# Document cleaning process
print("\nData cleaning completed. Data shape:", data_cleaned.shape)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Exploratory Analysis
# Visualizing distribution of ratings
plt.figure(figsize=(10, 6))
sns.histplot(data_cleaned['rating'], bins=30, kde=True)
plt.title('Distribution of Recipe Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()

# Visualizing relationship between calories and rating
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data_cleaned, x='calories', y='rating', alpha=0.6)
plt.title('Calories vs Rating')
plt.xlabel('Calories')
plt.ylabel('Rating')
plt.show()

# Visualizing top ingredients used in highly rated recipes
# Extracting ingredient columns
ingredient_columns = data_cleaned.columns[6:]
top_ingredients = data_cleaned[ingredient_columns].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_ingredients.index, y=top_ingredients.values)
plt.title('Top 10 Most Common Ingredients in Highly Rated Recipes')
plt.xticks(rotation=45)
plt.xlabel('Ingredients')
plt.ylabel('Count')
plt.show()

# Step 2: Detailed Insights
# 1. Common ingredients in highly rated recipes
common_ingredients = top_ingredients.index.tolist()
print("Common ingredients in highly rated recipes:", common_ingredients)

# 2. Correlation between preparation time and recipe ratings
# sns.scatterplot(data=data_cleaned, x='prep_time', y='rating')

# 3. Insights on improving user experience
print("\nInsights:")
print("1. The most common ingredients in highly rated recipes include:", common_ingredients)





```

```python
#Common Ingredients in Highly Rated Recipes
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv(r"C:\Users\nadha.NICHU\OneDrive\Desktop\epi_r.csv")

# Filter out highly rated recipes
high_rated_recipes = data[data['rating'] >= 4.0]

# Extracting ingredients and counting occurrences
ingredient_columns = high_rated_recipes.columns[30:]
ingredients_counts = high_rated_recipes[ingredient_columns].sum().sort_values(ascending=False).head(10)

# Plotting the top ingredients
plt.figure(figsize=(10, 6))
sns.barplot(x=ingredients_counts.index, y=ingredients_counts.values, palette='viridis')
plt.xticks(rotation=45)
plt.title('Top 10 Ingredients in Highly Rated Recipes')
plt.xlabel('Ingredients')
plt.ylabel('Count')
plt.show()
```

```python
# Scatter plot of calories vs. ratings
plt.figure(figsize=(10, 6))
sns.scatterplot(x='calories', y='rating', data=data, alpha=0.6)
plt.title('Calories vs. Recipe Ratings')
plt.xlabel('Calories')
plt.ylabel('Rating')
plt.xlim(0, data['calories'].max())
plt.ylim(0, 20)
plt.axhline(y=3.5, color='r', linestyle='--', label='Average Rating')
plt.legend()
plt.show()
```

```python
# Histogram of recipe ratings
plt.figure(figsize=(10, 6))
sns.histplot(data['rating'], bins=30, kde=True, color='blue', alpha=0.6)
plt.title('Distribution of Recipe Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.axvline(data['rating'].mean(), color='r', linestyle='--', label='Mean Rating')
plt.legend()
plt.show()
```

```python
# Boxplot to visualize outliers in ratings
plt.figure(figsize=(10, 6))
sns.boxplot(y='rating', data=data, color='cyan')
plt.title('Boxplot of Recipe Ratings')
plt.ylabel('Rating')
plt.axhline(y=3.5, color='r', linestyle='--', label='Average Rating')
plt.legend()
plt.show()
```
